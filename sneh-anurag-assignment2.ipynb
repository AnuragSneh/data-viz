```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/Users/anuragsneh/Desktop/uiuc_msim/data viz/building_inventory.csv")
```

```python
#solution1
df1=df.sort_values("Year Acquired")

plt.scatter(df1["Year Acquired"],df1["Year Constructed"])
plt.xlabel("Years Acquired")
plt.ylabel("Years Constructed")
plt.title("Years Acquired VS Years Constructed")
```

```python
#After Magnification into the dense data and ignoring the extreme outliers

df2=df1[(df1["Year Acquired"]>1775) & (df1["Year Constructed"]>1550)]
plt.scatter(df2["Year Acquired"],df2["Year Constructed"])
plt.xlabel("Year Acquired")
plt.ylabel("Year Constructed")
plt.title("Years Acquired VS Years Constructed")
```

1. There is a linear relationship between majority of the data points, barring some obvious outliers. 
These can be either when the construction date is before the acquired date or when the construction date 
has been extended by an abnormally long delay. I have initially ploted a normal plot and observed outliers 
and dense concentration of points at one particular time frame. After looking at the concentated data points we conclude that the relationship is linear. The current visualization suitably captures the relationship in between the variables. The only shortcoming to such visualization is that the scatter points should have been less opaque.



```python
#Solution-2
df_sum=df.groupby("Congress Dist").sum()
df_sum.head()
print(df_sum.loc[:,["Square Footage"]])
```

```python
#df_sum.plot()
df_sum["congress"]=df_sum.index
df_sum.plot.bar("congress", "Square Footage")
plt.xlabel("Congressional District")
plt.ylabel("Square Footage")
plt.title("Total square footage as a function of congressional district")
```

2. The graph gives us the visual representation of total square footage as function of Congressional District. 
The y axis is given to the scale of 10 to the power 7. The bar graph gives the total square footage. Congress 13 seems to have the maximum total square footage. The current graph conveys the prominent congress district in the best possible manner. Though, it could be better if could drawn a line graph as that would have shown the continuous trend. 


```python
# solution 3

df["square_footage_average"]=df["Square Footage"]/df["Total Floors"]
df1=df[df["Total Floors"]!=0]
df1_mean=df1.groupby("Congress Dist")["square_footage_average"].mean()
plt.plot(df1_mean.index,df1_mean)
plt.xlabel("Congress District")
plt.ylabel("Average Square footage per floor")
```

3. The above line graph shows the average square footage per floor. The various crests and troughs represents 
the average square footage per floor corresponing to the various congress district. 
Our current graph gives an holistic view of the upward and downward trends of the given dataset.
Though, A bar graph could have given us a better understanding of problem statement.
